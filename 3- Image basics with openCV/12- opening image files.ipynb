```python
import numpy as np 
import matplotlib.pyplot as plt  
%matplotlib inline 
```

```python
import cv2 
img = cv2.imread('../DATA/00-puppy.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

## how to read gray scale image 

```python
img_gray=cv2.imread('../DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray')
```

### rehsape the image

```python
img.shape
```

```python
img_new=cv2.resize(img,(1000,400))
```

```python
plt.imshow(img_new)
```

## flip the image 

```python
flip_img=cv2.flip(img,0)
plt.imshow(flip_img)
```

```python
cv2.imwrite('flip.jpg',flip_img)
```

```python
fig=plt.figure(figsize=(2,2))
ax=fig.add_subplot(111)
ax.imshow(flip_img)
```
